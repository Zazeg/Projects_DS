## Предобработка данных

1. В данных очень много пропусков, в половине столбцов они присутствуют.
2. Нужно изменить тип на int у столбца с количеством этажей, балконов, сколько дней было размещено объявление, расстояние до парка ближайшего, аэропорта, озера, количество парков, озер.
3. Апартаменты должен быть столбец булевый. Также в этом столбце много отсутствующих данных, их надо заменить на False, потому что квартир, как правило, больше чем апартаментов, и люди просто не заполняют эти данные.
4. Высоту потолков в отсутствующих данных можно заменить на среднюю высоту потолков по населенному пункту, а можно заменить на стандартную высоту потолков в панельных домах, которая равна 2.6 метра. Ниже этой высоты в жилых домах найти сложно, а если в действительности будет выше - ну что же, потенциальным покупателям бонус. Выберу второй вариант.
5. Всего этажей не указано лишь в небольшом количестве данных. Заменю эти данные на указанный этаж +1 этаж, чтобы не искажались исследования с данными квартир на последнем этаже.
6. Жилая площадь не указана в более чем 1900 данных. Может продавцы специально не указали эти данные, а может это не квартиры, а апартаменты, в которых нет жилой площади, потому что юридически это нежилые помещения. Заменить эти данные можно только на 0, общая площадь всё равно будет указана.
7. В датах публикации совершенно бесполезная информация о времени публикации. Надо изменить столбец на день, месяц и год.
8. Пропущенные данные с площадью кухни изменю на 0. Кстати, сейчас существуют кваритры, где люди специально не делают кухни. Питаться можно в ресторанах/кафе.
9. Пропущенные данные в балконах нужно сменить на 0
10. В 49 данных пропущено название населенного пункта. От этих данных лучше избавиться - на общую картину это вряд ли повлияет.
11. Расстояния до парков, озер и аэропорта некоторые люди не считают важным указывать. Заменю эти пропущенные данные на 0.
12. Количество дней от начала публикации до снятия. Пропущенные данные могут указывать на то, что эти данные до сих пор не сняты с продажи, например. Или и дня не прошло, а они уже продались. Заменю их на 0.

```python
realty['is_apartment'] = realty['is_apartment'].fillna(value=False) # Исправляю столбец is_apartment
realty['is_apartment'] = realty['is_apartment'].astype('bool') # Делаю его булевым

realty['ceiling_height'] = realty['ceiling_height'].fillna(value = 2.60) #Замена высоты потолков на 2.60

realty['floors_total'] = realty['floors_total'].fillna(value = realty['floor'] + 1) # Замена пропущенных значений числа этажей в доме
print('Информация о данных:')
print()
realty.info()
```

Итак, заменим оставшиеся пропущенные данные на 0, кроме данных с пропущенным названием населенного пункта.

```python
realty = realty.fillna(value=0)
realty = realty[realty['locality_name'] != 0]
```

```python
# Заменяем данные столбцов на int:
realty.astype({'floors_total': 'int32', 'balcony': 'int32', 'airports_nearest': 'int32',
           'cityCenters_nearest': 'int32', 'parks_around3000': 'int32', 'parks_nearest': 'int32',
           'ponds_around3000': 'int32', 'ponds_nearest': 'int32', 'days_exposition': 'int32'})
```

## Расчёты и добавление результатов в таблицу

Нет необходимости в данных указывать время размещения объявления, нужна только дата. Тем более, в первых и последних данных время указано одинаковое, 00:00:00. Но создам отдельный столбец с датой:

```python
realty['date_exposition'] = pd.to_datetime(realty['first_day_exposition'], format='%Y-%m-%d')

realty['weekday'] = realty['date_exposition'].dt.weekday # добавление стобца с номером дня недели
realty['month'] = realty['date_exposition'].dt.month # столбца с месяцем
realty['year'] = realty['date_exposition'].dt.year # и столбца с годом
# Добавление столбца с ценой квадратного метра и округление его до копеек
realty = realty.assign(meter_price = realty.last_price / realty.total_area).round(2) # Добавление столбца с ценой квадратного метра

# Добавление столбца с отношением жилой площади к общей
realty = realty.assign(living_total_ratio = realty.living_area / realty.total_area)

# Добавление столбца с отношением площади кухни к общей
realty = realty.assign(kitchen_total_ratio = realty.kitchen_area / realty.total_area)

# добавление столбца со значением этажа
def what_floor(row):
    """
    Функция принимает в качестве данных ряд и проверяет этаж.
    Является ли этаж первым, последним или другим.
    """
    floor = row['floor']
    total_floors = row['floors_total']
    if floor == 1:
        return 'первый'
    if floor == total_floors:
        return 'последний'
    return 'другой'

realty['what_floor'] = realty.apply(what_floor, axis=1)
```

### Изучение предложения квартир: для каждой квартиры есть информация о расстоянии до центра. Выделение квартир в Санкт-Петербурге (locality_name). Необходимо выяснить, какая область входит в центр. Создание столбца с расстоянием до центра в километрах: округление до целых значений. Расчет средней цены для каждого километра. Построение графика зависимости цены от удалённости от центра. Определение границы где график сильно меняется, — это и будет центральная зона.

```python
spb_flats = realty.query('locality_name == "Санкт-Петербург"')
spb_flats = spb_flats.query('cityCenters_nearest > 0')
spb_prices = spb_flats.loc[:, ['locality_name', 'cityCenters_nearest','last_price']]
spb_prices['cityCenters_nearest'] = (spb_prices['cityCenters_nearest'] / 1000).round() #расстояние до центра переводим в км
spb_prices.astype({'cityCenters_nearest': 'int32'})
```

Построю график зависимости стоимости м2 по расстоянию от центра

```python
spb_prices.pivot_table(index='cityCenters_nearest', values='last_price', aggfunc=('mean')).plot(y='last_price', kind='bar', figsize=(8, 8))
```

По графику видно, что самые дорогие цены за квадратный метр в самом центре, до 1 км, затем наблюдается резкое падение цены. Далее падение наблюдается после 2 км от центра, а также сильное падение цены после 7 км от центра.
