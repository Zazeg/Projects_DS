## Модель

### Написание функции для вычисления итоговой sMAPE.

```python
def smape(target, predictions):
    return 1 / len(target) * np.sum(np.abs(predictions - target) / ((np.abs(target) + np.abs(predictions)) / 2)) * 100
```

```python
def total_smape (smape_rougher, smape_final):
    total_smape = 0.25 * smape_rougher + 0.75 * smape_final
    return total_smape
```

### Обучение разных моделей и оценка их качества кросс-валидацией. Выбор лучшей модели и проверка её на тестовой выборке

#### Подготовка выборок

Для обучения моделей нужно подготовить выборки и выделить целевые признаки, а также масштабировать их (для модели линейной регрессии).

```python
rougher_target_train = data_train_model['rougher.output.recovery']
final_target_train = data_train_model['final.output.recovery']
features_train = data_train_model.drop(['date', 'rougher.output.recovery', 'final.output.recovery'], axis=1)

rougher_target_test = data_test_model['rougher.output.recovery']
final_target_test = data_test_model['final.output.recovery']
features_test = data_test_model.drop(['date', 'rougher.output.recovery', 'final.output.recovery'], axis=1)
```

```python
scaler = StandardScaler()
scaler.fit(features_train)
features_train_linear = scaler.transform(features_train)
features_test_linear = scaler.transform(features_test)
```

#### Модель линейной регрессии

```python
smape_score = make_scorer(smape, greater_is_better=False)

cross_val_score_rougher = cross_val_score(LinearRegression(), 
                                          features_train_linear, 
                                          rougher_target_train,
                                          cv=4, 
                                          scoring=smape_score).mean()
cross_val_score_final = cross_val_score(LinearRegression(),
                                        features_train_linear,
                                        final_target_train,
                                        cv=4, 
                                        scoring=smape_score).mean()
total_smape_linear = total_smape(cross_val_score_rougher, cross_val_score_final)
print('Итоговое sMAPE:', total_smape_linear)
```

#### Модель дерева решений

```python
cross_val_score_rougher = cross_val_score(DecisionTreeRegressor(random_state=12345),
                                          features_train,
                                          rougher_target_train,
                                          cv=4, 
                                          scoring=smape_score).mean()
cross_val_score_final = cross_val_score(DecisionTreeRegressor(random_state=12345),
                                        features_train,
                                        final_target_train,
                                        cv=4, 
                                        scoring=smape_score).mean()
total_smape_result = total_smape(cross_val_score_rougher, cross_val_score_final)
print('Итоговое sMAPE:', total_smape_result)
```

#### Модель случайного леса

```python
for depth in range(11, 13):
    
    for estim in range(3, 5):
        
        model = RandomForestRegressor(random_state=12345, 
                                       n_estimators=estim, 
                                       max_depth=depth)
        
        smape_score = make_scorer(smape, greater_is_better=False) 
        
        smape_best = -100
        best_depth = 1
        best_estim = 1
        
        cross_val_score_rougher = cross_val_score(model, features_train, rougher_target_train, cv=4, scoring=smape_score).mean()
        cross_val_score_final = cross_val_score(model, features_train, final_target_train, cv=4, scoring=smape_score).mean()

        total_smape_result = total_smape(cross_val_score_rougher, cross_val_score_final)
        if total_smape_result > smape_best:
            smape_best = total_smape_result
            best_depth = depth
            best_estim = estim
        
print('Итоговое sMAPE рандомного дерева:', smape_best,', depth =', best_depth,', estimator =', best_estim)
```

Модель случайного леса показала лучшее значение метрики sMAPE.

#### Проверка модели на тестовой выборке

```python
model_forest = RandomForestRegressor(random_state=12345, n_estimators=4, max_depth=12)

model_forest.fit(features_train, rougher_target_train)
predictions_rougher = model_forest.predict(features_test)

model_forest.fit(features_train, final_target_train)
predictions_final = model_forest.predict(features_test)

rougher_smape = smape(rougher_target_test, predictions_rougher)
final_smape = smape(final_target_test, predictions_final)

print('Итоговое sMAPE:', total_smape(rougher_smape, final_smape))
```

Сравнение нашей модели с константной моделью DummyRegressor и проверка на адеватность.
